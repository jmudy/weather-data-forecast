# ARIMA

```python
df["fecha"] = pd.to_datetime(df["fecha"], format="%Y-%m-%d")
df.set_index("fecha", inplace=True)
```

```python
df_agrupado_1M = df.groupby(pd.Grouper(freq="1M")).mean()
```

```python
ts = pd.Series(df_agrupado_1M["tmed"])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(ts.index, ts)
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.title('Serie temporal de la temperatura media')

plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

plot_acf(ts, lags=12, ax=ax1)
ax1.set_xlabel('Retardo')
ax1.set_ylabel('ACF')
ax1.set_title('Función de Autocorrelación (ACF) serie original')

plot_pacf(ts, lags=12, ax=ax2)
ax2.set_xlabel('Retardo')
ax2.set_ylabel('PACF')
ax2.set_title('Función de Autocorrelación Parcial (PACF) serie original')

plt.tight_layout()

plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result_adf = adfuller(ts)
print("P-valor antes de la diferenciación:", round(result_adf[1],4))
```

```python
ts_diff = ts.diff().dropna()  # Diferenciación de primer orden
result_adf_diff = adfuller(ts_diff)
print("Valor p después de la diferenciación:", round(result_adf_diff[1],4))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

plot_acf(ts_diff, lags=12, ax=ax1)
ax1.set_xlabel('Retardo')
ax1.set_ylabel('ACF')
ax1.set_title('Función de Autocorrelación (ACF) serie diferenciada')

plot_pacf(ts_diff, lags=12, ax=ax2)
ax2.set_xlabel('Retardo')
ax2.set_ylabel('PACF')
ax2.set_title('Función de Autocorrelación Parcial (PACF) serie diferenciada')

plt.tight_layout()

plt.show()
```
